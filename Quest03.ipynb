```python
import re
from collections import Counter

# 1. 파일을 읽어서 내용을 가져온다. 2. 변수에 초기화
with open('Avengers.txt', 'r', encoding='utf-8') as file:
  script = file.read()
```

```python
# 모든 문자를 소문자로 바꿔준다.
script = script.lower()
```

```python
# 모든 기호를 제거해준다.
script = re.sub(r"[^\w\s]", ' ', script)
```

```python
# 파일의 내용을 단어 단위로 잘라준다.
words = script.split()
```

```python
# 2-gram을 생성한다.
two_gram = zip(words, words[1:])
```

```python
print(two_gram)
```

```python
# Counter 내장함수로 빈도수를 계산한다.
two_gram_cnt = Counter(two_gram)
```

```python
print(two_gram_cnt)
```

```python
# most_common 내장함수를 사용하여 빈도가 높은 것을 찾는다.
max2gram = two_gram_cnt.most_common(1)
```

```python
print(max2gram)
print(two_gram_cnt)
```
